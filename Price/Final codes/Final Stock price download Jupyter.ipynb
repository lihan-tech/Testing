```python
from fyers_apiv3 import fyersModel
from datetime import datetime
import pandas as pd
import pytz
import os
import time
import requests

# Function to check internet connection
def is_internet_available():
    try:
        requests.get("https://www.google.com", timeout=5)
        return True
    except (requests.ConnectionError, requests.Timeout):
        return False

# Wait for internet connection
def wait_for_internet():
    while not is_internet_available():
        print("No internet connection. Waiting for internet connection...")
        time.sleep(5)

# Use the provided token directly
access_token = "eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJpc3MiOiJhcGkuZnllcnMuaW4iLCJpYXQiOjE3MjUyNzIyNjMsImV4cCI6MTcyNTMyMzQ0MywibmJmIjoxNzI1MjcyMjYzLCJhdWQiOlsieDowIiwieDoxIiwieDoyIiwiZDoxIiwiZDoyIiwieDoxIiwieDowIl0sInN1YiI6ImFjY2Vzc190b2tlbiIsImF0X2hhc2giOiJnQUFBQUFCbTFaREgyS21qSlEzV3kzS19WdXEtMnYyOTZPNEFyVWRmQ3E2TjNqVmI1Y3hZUGlNSDFpQlFzblZiWDZBQ2dzLXpqUE1EbnNwV0l0Vzc1aHh2OEV4R3BoVlRYaEEtM25UcTl3SGNxdDdyWFRBeEh1OD0iLCJkaXNwbGF5X25hbWUiOiJIVUxJS0FUVEUgTkFHQVJBSkFQUEEgUkFTSE1JIiwib21zIjoiSzEiLCJoc21fa2V5IjoiZDg1OGE2OTQyNTJiN2UyZmM2MGM2ZTc0NWVhNzg3Mjc2YWI2OTc4OGExOGExYzY5MWVkMDU3ZGMiLCJmeV9pZCI6IllIMDAzMDIiLCJhcHBUeXBlIjoxMDAsInBvYV9mbGFnIjoiTiJ9.7f3jzcS5InwBL-9wrFlDxKpXb4YLhHYYQfUYRUviTjk"

client_id = "D7D574OOFI-100"
fyers = fyersModel.FyersModel(client_id=client_id, token=access_token, log_path=r'C:\Users\bomma\Desktop\Daily Reports\Price')

ist = pytz.timezone('Asia/Kolkata')

# Function to fetch stock data
def fetch_stock_data(symbol):
    max_retries = 7  # Increased max retries
    range_from = "2024-09-02"  # Start date of the range
    range_to = "2024-09-02"    # End date of the range
    for attempt in range(max_retries):
        try:
            data = {
                "symbol": f"NSE:{symbol}",
                "resolution": "5",
                "date_format": "1",
                "range_from": range_from,  # Use the specified start date
                "range_to": range_to,      # Use the specified end date
                "cont_flag": "1"
            }
            response = fyers.history(data=data)

            if 'candles' not in response:
                print(f"No candles data returned for {symbol}. Response: {response}")
                return pd.DataFrame(), False  # Return a flag indicating no candles data

            formatted_candles = [
                [datetime.utcfromtimestamp(candle[0]).replace(tzinfo=pytz.utc).astimezone(ist).strftime('%Y-%m-%d %H:%M:%S'),
                 candle[1], candle[2], candle[3], candle[4], candle[5]]
                for candle in response['candles']
            ]
            return pd.DataFrame(formatted_candles, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume']), True
        except Exception as e:
            if 'code' in response and response['code'] == 429:
                wait_time = 2 ** (attempt + 1)  # More aggressive exponential backoff
                print(f"Rate limit reached. Retrying in {wait_time} seconds...")
                time.sleep(wait_time)
            else:
                print(f"Error occurred while fetching data for {symbol}: {e}")
                return pd.DataFrame(), False

# Specify the Excel file path for symbols
excel_file_path = r'C:\Users\bomma\Desktop\Daily Reports\Price\Stocks symbol\Symbols 2022.xlsx'

# Load symbols from Excel file
try:
    symbols_df = pd.read_excel(excel_file_path, sheet_name='Rama')
except Exception as e:
    print(f"Error reading the symbols Excel file: {e}")
    symbols_df = None

if symbols_df is not None and 'Symbols' in symbols_df.columns:
    # Extract symbols from the 'Symbols' column
    symbol_list = symbols_df['Symbols'].tolist()

    # Specify the folder path
    output_folder_path = r'C:\Users\bomma\Desktop\Daily Reports\Price\Day wise'
    no_symbols_file_path = r'C:\Users\bomma\Desktop\Daily Reports\Price\No download\No symbols.xlsx'
    error_log_file_path = r'C:\Users\bomma\Desktop\Daily Reports\Price\error\Error log.xlsx'

    # Ensure the output folder exists
    os.makedirs(output_folder_path, exist_ok=True)

    no_symbols = []
    error_symbols = []
    successful_symbols = []

    batch_size = 51  # Process symbols in batches to help with rate limiting
    for i in range(0, len(symbol_list), batch_size):
        batch = symbol_list[i:i + batch_size]
        for index, stock_symbol in enumerate(batch, start=1):
            # Check internet connection before fetching data
            wait_for_internet()

            df_stock, has_candles = fetch_stock_data(stock_symbol)

            if not has_candles:
                no_symbols.append(stock_symbol)
                continue

            if df_stock.empty:
                error_symbols.append(stock_symbol)
                continue

            try:
                # Write the data to a CSV file
                csv_file_path = os.path.join(output_folder_path, f'{stock_symbol}_data.csv')
                df_stock.to_csv(csv_file_path, index=False)
                print(f"CSV file {i + index}/{len(symbol_list)} saved for symbol {stock_symbol}")
                successful_symbols.append(stock_symbol)
            except Exception as e:
                print(f"Error occurred while saving CSV for {stock_symbol}: {e}")
                error_symbols.append(stock_symbol)

        # Wait between batches to avoid hitting rate limits
        print(f"Waiting 60 seconds before processing the next batch...")
        time.sleep(20)

    # Save the symbols with no candles data to an Excel file
    if no_symbols:
        try:
            no_symbols_df = pd.DataFrame(no_symbols, columns=['Symbols'])
            no_symbols_df.to_excel(no_symbols_file_path, index=False)
            print(f"Symbols with no candles data saved to '{no_symbols_file_path}'")
        except Exception as e:
            print(f"Error occurred while saving no symbols data: {e}")

    # Save the symbols with other errors to an Excel file
    if error_symbols:
        try:
            error_symbols_df = pd.DataFrame(error_symbols, columns=['Symbols'])
            error_symbols_df.to_excel(error_log_file_path, index=False)
            print(f"Symbols with errors saved to '{error_log_file_path}'")
        except Exception as e:
            print(f"Error occurred while saving error symbols data: {e}")

    print(f"All CSV files saved to '{output_folder_path}'.")
else:
    print("Symbols data not found or 'Symbols' column missing in the Excel file.")
```

```python
import os
import sqlite3
import csv

# Source folder path
source_folder = r"C:\Users\bomma\Desktop\Daily Reports\Price\Day wise"

# Destination database path
destination_db = r"C:\Users\bomma\Desktop\Daily Reports\Price\Final Price database\Sorting 2022_2023.db"

# Connect to the destination database
conn = sqlite3.connect(destination_db)
cursor = conn.cursor()

# Iterate through each CSV file in the source folder
for file_name in os.listdir(source_folder):
    if file_name.endswith(".csv"):
        table_name = os.path.splitext(file_name)[0]  # Table name is the file name without extension
        csv_file_path = os.path.join(source_folder, file_name)
        
        # Check if table exists in the destination database
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name=?", (table_name,))
        table_exists = cursor.fetchone()
        
        if table_exists:
            # Read data from CSV file
            with open(csv_file_path, 'r') as file:
                csv_reader = csv.reader(file)
                headers = next(csv_reader)
                placeholders = ", ".join(['?' for _ in headers])
                
                # Insert data into the table
                insert_query = f"INSERT INTO {table_name} VALUES ({placeholders})"
                try:
                    for row in csv_reader:
                        cursor.execute(insert_query, row)
                    conn.commit()
                    print(f"Data from {file_name} inserted into table {table_name} successfully.")
                except sqlite3.Error as e:
                    print(f"Error inserting data into table {table_name}: {e}")
                    conn.rollback()
        else:
            print(f"Table {table_name} does not exist in the destination database. Skipping...")
            

# Close connection to the database
conn.close()
```
